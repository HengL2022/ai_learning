# Chapter 2 — if-else and Flow Control

This chapter introduces decision-making in programs: how to evaluate conditions and branch execution using `if`, `elif`, and `else` statements.

## 2.1 Boolean Values

- The Boolean type has exactly two values: `True` and `False` (capitalized, no quotes).
- `True` and `False` cannot be used as variable names.

## 2.2 Comparison Operators

Comparison operators compare two values and return a Boolean:

- `==` (equal to), `!=` (not equal to)
- `<` (less than), `>` (greater than)
- `<=` (less than or equal), `>=` (greater than or equal)

**Important distinction:** `==` asks "are these equal?" while `=` assigns a value to a variable.

- An integer and a float can be equal (`42 == 42.0` is `True`), but an int never equals a string (`42 == '42'` is `False`).

````python
# equal to 
print('a' == 'b')
print('cc' == 'cc')
````

````python
# not equal to 
print('a' != 'b')
print('cc' != 'cc')
````

````python
# less/greater than
print(6 < 5)
print(6 > 3)
````

````python
# less/greater or equal than
print(6 <= 6)
print(6 >= 3)
````

## 2.3 Boolean Operators

- `and` — `True` only if both operands are `True`.
- `or` — `True` if at least one operand is `True`.
- `not` — inverts a single Boolean value.

**Evaluation order:** `not` is evaluated first, then `and`, then `or` (after all math and comparisons).

Boolean and comparison operators can be combined in complex expressions:

````python
(8 < 9) or (8 > 9)
````

````python
not (8 > 9)
````

````python
(8 < 9) and (8 > 9)
````

## 2.4 Conditions and Blocks of Code

- A condition is an expression that evaluates to `True` or `False`.
- A block is a group of indented lines that begins when indentation increases and ends when it decreases.
- A colon (`:`) always precedes a new block.
- Blocks can be nested inside other blocks.

````python
age = 19
if age > 18:        # colon here — starts a block
    print("adult")  # this line is INSIDE the block (indented)
    print("can vote")  # also inside

print("done")  # back OUTSIDE the block (no indent)
````

````python
def check_students(students):       # level 1
    for name in students:           # level 2 (nested in function)
        if name == "Alice":         # level 3 (nested in for loop)
            print("Found Alice!")   # level 4 (nested in if)
````

## 2.5 Flow Control Statements

### if Statements

- Executes the clause only if the condition is `True`; otherwise skips it.

### else Statements

- Executes when the `if` condition is `False` — provides a fallback path.

### elif Statements

- Short for "else if" — checked only if all previous conditions were `False`.
- Order matters: once a `True` condition is found, remaining `elif`/`else` blocks are skipped.
- Only one clause in an `if`/`elif`/`else` chain will ever execute.

**Pattern:** `if` → zero or more `elif` → optional `else` (guarantees at least one clause executes).

````python
name = input("Who are you?")

if name == 'Heng':
    print('Hi, Heng!')
elif name == 'Max':
    print('Hi, Max!')
elif name == 'Mia':
    print('Hi, Mia!')
else:
    print('You must be Liz!')
````

# Chapter 3 — Loops

Loops let you repeat code automatically — the core of what makes programming powerful for automation. This chapter covers `while` loops, `for` loops, and how to control loop execution.

## 3.1 while Loops

- Repeats a block of code as long as the condition remains `True`.
- The condition is re-checked at the start of each iteration.

**Key difference from if:** An `if` clause runs once; a `while` clause loops back and re-checks.

````python
spam = 0
while spam<5:
    print ('Hello, world!')
    spam = spam + 1
````

## 3.2 break Statements

- Immediately exits the current loop, regardless of the condition.
- Commonly used with `while True:` (infinite loop) to create loops that exit from the middle.

````python
while True:
    name = input('>')
    if name == 'your name':
        break
````

## 3.3 continue Statements

- Jumps back to the start of the loop and re-evaluates the condition.
- Skips the remaining code in the current iteration.
- Both `break` and `continue` can only be used inside `while` and `for` loops.

````python
i = 0
while i < 10:
    i += 1
    if i == 4:
        continue  # skip printing when i is 4
    if i == 6:
        continue
    print (i)
````

## 3.4 Truthy and Falsey Values

- In conditions, `0`, `0.0`, and `''` (empty string) are treated as `False`.
- All other values are treated as `True`.
- The `bool()` function reveals a value's truthiness.

````python
bool (0)
````

````python
bool ('')
````

````python
bool ('5.0')
````

## 3.5 for Loops and range()

- `for` loops execute a block a specific number of times.
- `range(n)` produces the sequence 0, 1, 2, …, n-1 (up to but not including n).

````python
for i in range(20):
    print(i) 
````

### range() with Multiple Arguments

- `range(start, stop)` — starts at `start`, ends before `stop`.
- `range(start, stop, step)` — adds `step` between each iteration.
- Negative step counts down: `range(5, -1, -1)` → 5, 4, 3, 2, 1, 0.

**Why "up to but not including"?** This "closed, open" convention makes range size = stop − start and avoids off-by-one errors.

````python
for i in range (20, -20, -4):
    print(i)
````

## 3.6 Importing Modules

- Modules are Python files containing related functions (e.g., `random`, `math`, `sys`, `os`).
- Import with: `import random` (then call `random.randint(1, 10)`).
- Alternative: `from random import *` (no prefix needed, but less readable).

**Warning:** Never name your files after Python modules (e.g., `random.py`) — this causes import conflicts.


````python
# Method 1
import random

print(random.randint(1, 10)) #pick a random number between 1 and 10
print(random.choice(['a', 'b', 'c']))

#You always prefix the function with the module name (random.). This makes it crystal clear where the function came from.
````

````python
# Method 2
from random import *

print(randint(1, 10))
print(choice(['a', 'b', 'c']))
# * means import everything, so you do not need prefix, but this can have comfusion where they came from
````

````python
# Method 3
from random import randint
print(randint(1, 10))
````

### ⚠️ The naming conflict trap

This is a subtle but important warning. When you write `import random`, Python searches for a file named `random.py` — and it **checks your own folder first**.
```
my_project/
├── random.py        ← uh oh, you named your file this
└── main.py

python# Inside main.py
import random        # Python finds YOUR random.py, not the built-in one
random.randint(1,10) # AttributeError — your file doesn't have randint!

Your file shadows the real module. Python never even reaches the built-in. The fix is simple: never name your files after standard modules. Name it my_random.py, dice.py, anything else.

# Chapter 4 — Functions

Functions let you organize code into reusable, self-contained blocks. This chapter covers defining functions, parameters, return values, scope, and exception handling.

## 4.1 Defining and Calling Functions

- Use the `def` keyword to define a function: `def hello():`
- The function body is an indented block after the `def` statement.
- Call a function by its name followed by parentheses: `hello()`

**Key benefit:** Functions eliminate duplicated code (deduplication) — write once, call many times.

````python
def hello():
    print ('Hello, how are you!')

hello()
````

## 4.2 Arguments and Parameters

- Arguments are values passed in a function call: `say_hello_to('Alice')`.
- Parameters are the variables inside the function definition that receive those arguments.
- Parameters are destroyed when the function returns — they exist only in the function's local scope.

**Terminology:** define a function, call it, pass arguments into parameters.

````python
def say_hello_to(a):
    print ('hello!' + a)

print (say_hello_to('Alice'))
print (a)
````

## 4.3 Return Values and return Statements

- A `return` statement specifies the value a function sends back to the caller.
- The returned value can be stored in a variable or used directly in an expression.
- A function call in an expression evaluates to its return value.

````python
# Basic Ideal

def jia(a, b):
    return a + b
jia (3, 4)
````

When Python hits return, two things happen:

- The function stops executing immediately
- It hands the value back to wherever it was called from

````python
#Storing the return value in a variable
result = jia (3, 5)
print (result)
````

Using the return value directly in an expression
- You don't have to store it first — you can use the return value inline:

````python
print (jia(3, 6) *2 )
````

"A function call evaluates to its return value"
- This is the key mental model. Wherever you write a function call, Python replaces it with the return value:

````python
# What if there's no return?
# Python implicitly returns None — a special value meaning "nothing." This is fine for functions that do something (like print), but not for functions meant to calculate something.

def greet(name):
    print("Hello, " + name)   # does something, but returns nothing

result = greet("Alice")
print(result)    # None
````

````python
#return also exits the function immediately
def check(x):
    if x > 0:
        return "positive"   # exits here if x > 0
    return "not positive"   # only reached if x <= 0
````

## 4.4 The None Value

- `None` represents the absence of a value (`NoneType`).
- Functions without an explicit `return` statement return `None` implicitly.
- `print()` returns `None` — it displays text but has no meaningful return value.

## 4.5 Named (Keyword) Parameters

- Named parameters are identified by name rather than position in a function call.
- `print()` has two optional named parameters:
    - `end` — string appended after output (default: newline `'\n'`).
    - `sep` — separator between multiple arguments (default: space `' '`).


````python
print('Hello', end = '')
print('How are you?')
print('a','b','c', sep = ',')
````

## 4.6 The Call Stack

- Python uses a call stack to track which function is running and where to return when it finishes.
- Each function call creates a frame object on the stack.
- When a function returns, its frame is removed and execution resumes at the caller.
- Functions can call other functions, creating nested frames.

**Mental model:** Like a stack of plates — items are added and removed only from the top.

## 4.7 Local and Global Scope

- **Local scope:** variables created inside a function; exist only during that function call.
- **Global scope:** variables created outside all functions; accessible from anywhere.

### Scope Rules

1. Global code cannot access local variables.
2. One function's local scope cannot see another function's local variables.
3. Local code can read global variables.
4. Local and global variables can share names (but this is a bad practice).

**Why scoping matters:** It limits the "blast radius" of bugs — if a local variable has a bad value, you only need to search within that function.

## 4.8 The global Statement

- To modify a global variable from inside a function, declare it with `global`:

````python
def spam():
    global eggs
    eggs = 'spam'
````

````python
spam()
print (eggs)
````

## 4.9 Exception Handling (try / except)

- Errors (exceptions) can crash a program; `try`/`except` lets you handle them gracefully.
- Code that might fail goes in the `try` block; recovery code goes in `except`.

````python
try:
    result = int('not_a_number')
except ValueError:
    print('Invalid input!')
````
